```python
# Create a filtered dataframe from specific columns
premises_cols= ["country","province","region_1","winery","designation","variety","title","points","price","description",]
wine_130_ts= wine_data_data130_df[premises_cols].copy()

# Rename the column headers
wine_130_ts=wine_130_ts.rename(columns={"description":"review",
                                        "designation":"vineyard",
                                        "region_1":"region"
                                       })


wine_130_ts
```

```python
# Create a filtered dataframe from specific columns
premises_cols2= ["country","province","region_1","winery","designation","variety","points","price","description"]
wine_150_ts= wine_data_data150_df[premises_cols2].copy()

# Rename the column headers
wine_150_ts=wine_150_ts.rename(columns={"description":"review",
                                        "designation":"vineyard",
                                        "region_1":"region"
                                       })

wine_150_ts
```

```python
#merged tables
merged_df=wine_130_ts.merge(wine_150_ts, how="outer")

pd.set_option('display.max_columns', None)
merged_df
```

```python
#Looking for duplicates
duplicates = merged_df.loc[merged_df.duplicated(['review','country'])].sort_values('country')
duplicates
```

```python
#Droping Duplicates
clean_wine_db = merged_df.drop_duplicates()
clean_wine_db
```

```python
clean_wine_db.country.unique()
```

```python
#taking NA from region
clean_wine_db = clean_wine_db[clean_wine_db['region'].notna()]
clean_wine_db
```

```python
#taking NA from price
clean_wine_db = clean_wine_db[clean_wine_db['price'].notna()]
clean_wine_db
```

### GENERAL TABLE WITHOUT NORMALIZATION

```python
#taking NA from vineyard
clean_wine_db = clean_wine_db[clean_wine_db['vineyard'].notna()]
clean_wine_db1=clean_wine_db
clean_wine_db1
```

# 1st Table: COUNTRY

```python
#Country Catalogue
country= clean_wine_db["country"].sort_values().unique()

country = pd.DataFrame(country).reset_index()
country["index"] = country["index"]+1
country = country.rename(columns={0:"country"})
#country=country.set_index("index")
country
```

# 2nd Table: PROVINCE

```python
#Province Catalogue
province= clean_wine_db["province"].sort_values().unique()

province = pd.DataFrame(province).reset_index()
province["index"] = province["index"]+1
province = province.rename(columns={0:"province"})
#province=province.set_index("index")
province
```

# 3rd Table: REGION

```python
# 1st Table: COUNTRY#Region Catalogue
region= clean_wine_db["region"].sort_values().unique()

region = pd.DataFrame(region).reset_index()
region["index"] = region["index"]+1
region = region.rename(columns={0:"region"})
#region=region.set_index("index")
region
```

# 4th Table: WINERY

```python
# 1st Table: COUNTRY#Winery Catalogue
winery= clean_wine_db["winery"].sort_values().unique()

winery = pd.DataFrame(winery).reset_index()
winery["index"] = winery["index"]+1
winery = winery.rename(columns={0:"winery"})
#winery=winery.set_index("index")
winery
```

# 5th Table: VARIETY

```python
# 1st Table: COUNTRY#Variety Catalogue
variety= clean_wine_db["variety"].sort_values().unique()

variety = pd.DataFrame(variety).reset_index()
variety["index"] = variety["index"]+1
variety= variety.rename(columns={0:"variety"})
#variety=variety.set_index("index")
variety
```

# 6th Table: VINEYARD

```python
# 1st Table: COUNTRY#Vineyard Catalogue
vineyard= clean_wine_db["vineyard"].sort_values().unique()

vineyard = pd.DataFrame(vineyard).reset_index()
vineyard["index"] = vineyard["index"]+1
vineyard= vineyard.rename(columns={0:"vineyard"})
#vineyard=vineyard.set_index("index")
vineyard
```

# 7th Table: NORMALIZED DATA

```python
# 1st Table: COUNTRY# Change values to id's to insert in the db
clean_wine_db_df = clean_wine_db.merge(country, how="inner", left_on="country", right_on="country").rename(columns={"index": "country_id"}).drop(columns="country")
clean_wine_db_df = clean_wine_db_df.merge(region, how="inner", left_on="region", right_on="region").rename(columns={"index": "region_id"}).drop(columns="region")
clean_wine_db_df = clean_wine_db_df.merge(province, how="inner", left_on="province", right_on="province").rename(columns={"index": "province_id"}).drop(columns="province")
clean_wine_db_df = clean_wine_db_df.merge(winery, how="inner", left_on="winery", right_on="winery").rename(columns={"index": "winery_id"}).drop(columns="winery")
clean_wine_db_df = clean_wine_db_df.merge(variety, how="inner", left_on="variety", right_on="variety").rename(columns={"index": "variety_id"}).drop(columns="variety")
clean_wine_db_df = clean_wine_db_df.merge(vineyard, how="inner", left_on="vineyard", right_on="vineyard").rename(columns={"index": "vineyard_id"}).drop(columns="vineyard")

# Ordering Columns
premises_cols=["country_id","region_id","province_id","winery_id","variety_id","vineyard_id","title",
              "points","price","review"]
clean_wine_db_df= clean_wine_db_df[premises_cols].copy()
clean_wine_db_df
```
