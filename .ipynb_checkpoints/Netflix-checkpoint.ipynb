```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file1 = "Resources/netflix_titles.csv"
csv_file2 = "Resources/netflix-stocks-data.csv"
netflix_titles = pd.read_csv(csv_file1)
netflix_stocks = pd.read_csv(csv_file2)
netflix_titles.head()
```

```python
netflix_stocks.head()
```

```python
# Convert to date column of title dataset
def change_date(x):
    if type(x) in [int, float]:
        # If they don't contain date, put distant past date
        return "January 01 1970"
    s = x.replace(',', ' ').split()
    return f'{s[0]} {int(s[1]):2} {s[2]}'

netflix_titles['date_added'] = netflix_titles['date_added'].apply(change_date)

netflix_titles['date_added'] = pd.to_datetime(netflix_titles['date_added'], format="%B %d %Y")
netflix_titles.head()
```

```python
# Convert date format of stocks dataset
netflix_stocks['Date'] = pd.to_datetime(netflix_stocks['Date'], format="%m/%d/%Y")
netflix_stocks.head()
```
